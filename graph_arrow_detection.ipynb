## on multiple images

```python
import os
from pathlib import Path
from PIL import Image
from matplotlib import pyplot as plt

# Assuming detect.py is in the same directory as this script
detect_script_path = '/content/drive/MyDrive/graph_arrow_detection_2/yolov7/detect.py'

# Path to the YOLOR weights file
weights_path = '/content/drive/MyDrive/graph_arrow_detection_2/yolov7/runs/train/exp3/weights/best.pt'

# Path to the test images folder
test_images_folder = '/content/drive/MyDrive/graph_arrow_detection_2/yolov7/test/'

# List all test images
test_images = [img for img in os.listdir(test_images_folder) if img.lower().endswith((".png", ".jpg"))]

# Run YOLOR detection on each image
for i, img_name in enumerate(test_images):
    print('************************************************')
    print(i)
    print('************************************************')
    print('img_name=', img_name)
    image_path = os.path.join(test_images_folder, img_name)
    print('************************************************')
    print('image path=', image_path)

    # Run YOLOR detection using the detect.py script
    !python {detect_script_path} --weights {weights_path} --source {image_path}

    # Determine the corresponding output folder for the predicted image
    output_folder = f'/content/drive/MyDrive/graph_arrow_detection_2/yolov7/runs/detect/exp'
    if i > 0:
        output_folder += f'{i + 1}'
    print('************************************************')
    print('output folder=', output_folder)

    # Determine the corresponding output file extension based on the original image format
    print('************************************************')
    output_file_extension = os.path.splitext(img_name)[-1]
    print('out extension=', output_file_extension)

    # Load the original and predicted images
    original_image = Image.open(image_path).convert("RGB")
    predicted_image_path = os.path.join(output_folder, f"{img_name[:-4]}{output_file_extension}")
    print('************************************************')
    print('predicted image path=', predicted_image_path)
    predicted_image = Image.open(predicted_image_path).convert("RGB")

    # Display the images side by side
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))

    axes[0].imshow(original_image)
    axes[0].set_title("Original Image")
    axes[0].axis("off")

    axes[1].imshow(predicted_image)
    axes[1].set_title("Predicted Image")
    axes[1].axis("off")

    plt.show()
```
